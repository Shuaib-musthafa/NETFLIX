```python
sns.set(style="whitegrid")
plt.figure(figsize=(20,8))
sns.barplot(x='release_year', y='count', hue='type', data=data_counts)
plt.xticks(rotation=45)
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.title('Distribution of Content Categories/Types Over Release Years')
plt.legend(title='Category/Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
release_year = net['release_year']
# category_type = df['category_type']

# Filter the dataset to include only data from the year 2021
net_2021 = net[net['release_year'] == 2021]

# Count the occurrences of each category/type of content
category_counts = net_2021['type'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(category_counts, labels=category_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Content Categories/Types in 2021')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

# Show the plot
plt.show()
```

```python
movies_net = net[net['type'] == 'Movie']

# Group the data by release year and movie type
movie_counts = movies_net.groupby(['release_year']).size()

# Filter the data up to the year 2021
movie_counts_2021 = movie_counts.loc[:2021]

# Plot the line chart
plt.figure(figsize=(10, 6))
movie_counts_2021.plot(marker='o')
plt.title('Number of Movies Released Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.grid(False)

# Show the plot
plt.show()
```

```python
TV_Show_net = net[net['type'] == 'TV Show']

# Group the data by release year and TV Show type
TV_Show_counts = TV_Show_net.groupby(['release_year']).size()

# Filter the data up to the year 2021
TV_Show_counts_2021 = TV_Show_counts.loc[:2021]

# Plot the line chart
plt.figure(figsize=(10, 6))
TV_Show_counts_2021.plot(marker='o')
plt.title('Number of TV Show Released Over Time')
plt.xlabel('Release Year')
plt.ylabel('Number of TV Show')
plt.grid(False)

# Show the plot
plt.show()
```

```python
type = net['type']
release_year = net['release_year']
durations = net['duration']

# Filter the DataFrame to include only movies from 2021
movies_2021_net = net[(net['type'] == 'Movie') & (net['release_year'] == 2021)]

# Convert 'duration' column to numeric data type
durations_numeric = pd.to_numeric(movies_2021_net['duration'].str.extract('(\d+)')[0], errors='coerce')

# Remove missing or invalid values
durations_numeric = durations_numeric.dropna()

# Plot histogram for movie durations in 2021
plt.figure(figsize=(10, 6))
plt.hist(durations_numeric, bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Movie Durations in 2021')
plt.xlabel('Duration (minutes)')
plt.ylabel('Frequency')
plt.grid(True)

# Show the plot
plt.show()
```

```python
release_years = net['release_year']

# Plot histogram for release years
plt.figure(figsize=(10, 6))
plt.hist(release_years, bins=range(min(release_years), max(release_years) + 1), color='salmon', edgecolor='black')
plt.title('Distribution of Release Years')
plt.xlabel('Release Year')
plt.ylabel('Frequency')
plt.grid(False)

# Show the plot
plt.show()
```

```python
net['rating'] = pd.to_numeric(net['rating'], errors='coerce')

# Convert 'duration' column to numeric data type
net['duration'] = pd.to_numeric(net['duration'], errors='coerce')

# Drop rows with missing values in either 'user_rating' or 'duration' columns
net = net.dropna(subset=['rating', 'duration'])

# Plot scatter plot for user ratings vs. movie duration
plt.figure(figsize=(10, 6))
plt.scatter(net['duration'], net['rating'], alpha=0.5, color='blue')  # alpha parameter controls transparency
plt.title('User Ratings vs. Movie Duration')
plt.xlabel('Movie Duration (minutes)')
plt.ylabel('User Rating')
plt.grid(True)

# Show the plot
plt.show()
```

```python
sns.set(style="whitegrid")

# Create the bar plot
plt.figure(figsize=(12, 6))
barplot = sns.barplot(x='country', y='count', data=count)

# Set the color of the bars
for i, bar in enumerate(barplot.patches):
    if i < 3:  # First three bars
        bar.set_color('red')
    else:  # Remaining bars
        bar.set_color('black')

plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Count')
plt.title('Top 10 Countries using Netflix')
plt.tight_layout()
plt.show()
```
